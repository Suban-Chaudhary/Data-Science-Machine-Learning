```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import mode as mode
%matplotlib qt
%matplotlib inline
```

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df
```

```python
df = df.head()
```

```python
sns.barplot(x='lunch', y = 'writing score',estimator = np.mean,palette = 'coolwarm',data = df);
plt.show()
```

```python
sns.catplot(x = 'parental level of education', y = 'writing score', estimator = np.median, data = df)
plt.show()
```

```python
sns.regplot(x = df['reading score'], y = df['writing score'], fit_reg = False)
plt.show()
```

```python
sns.pairplot(df, palette = 'coolwarm')
plt.show()
```

```python
sns.lineplot(x = 'reading score', y = 'math score', estimator = np.median, hue = 'parental level of education',data = df)
plt.show()
```

```python
sns.relplot(x = 'reading score', y = 'math score', estimator = np.median, hue = 'parental level of education',data = df)
plt.show()
```

```python
sns.scatterplot(x = 'reading score', y = 'writing score', hue = 'math score', style = 'math score', data = df)
plt.show()
```

```python
sns.barplot(x = 'math score', y = 'writing score', hue = 'reading score', data = df)
plt.show()
```
